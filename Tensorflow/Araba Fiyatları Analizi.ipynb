```python
dataFrame = yeniDataFrame
dataFrame.describe()
```

```python
dataFrame=dataFrame[dataFrame.year !=1970]
dataFrame.groupby("year").mean()["price"]
```

```python
# transmission kaldırılmalı çünkü nümerik bir bilgi değil
dataFrame.head()
```

```python
dataFrame= dataFrame.drop("transmission",axis=1)
dataFrame.head()
```

```python
y=dataFrame["price"].values # price bizim outputumuz olacak. Numpy dizisine dönüştü
x=dataFrame.drop("price",axis=1).values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() # bu sınıftan bir obje oluşturduk
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x_train.shape
```

```python
model = Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(1)) # çıkış nöronu
model.compile(optimizer="adam",loss="mse")
```

```python
# x_test ve y_test'i tuple içinde verip manuel işlemi direkt yaparız. Test verilerini kullanımı için yazılan kod kolaylaşır
# batch size ile verileri kaçar kaçar vereceğimizi belirliyoruz. Veri büyük olduğu için yapıyoruz
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300) 
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
tahminDizisi=model.predict(x_test)
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
