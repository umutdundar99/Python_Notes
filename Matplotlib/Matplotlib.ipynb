```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# yaşa göre kilonun değişimi
yasListesi=[10,20,30,30,30,40,50,60,70]
kiloListesi=[20,60,70,80,95,65,90,86,63]
numpyYasListesi=np.array(yasListesi)
numpyKiloListesi=np.array(kiloListesi)
#"g" demek yeşil bir çizgiyle göster demek
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
# plt.show() başka bir derleyicide bu kodu yazdığımızda grafiği görüntüleyebilirdik
```

```python
numpyDizisi1=np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**3
numpyDizisi2
```

```python
# "g--" iki tire çizgi yazıldığında grafik kesikli olur
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
```

aynı düzlemde 2 farklı grafik çizimi

```python
# kaç sıra olduğu, kaç grafik olduğu, hangisi çizilecek
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2) # bir sıra olacak, iki grafik olacak, 2. grafiği çizdir
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
# eksen isimleri ve genişliğini belirtmek için kullanılacak
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Grafik Başlığı")
```

```python
benimFigure2 = plt.figure()
# eksen isimleri ve genişliğini belirtmek için kullanılacak
# son iki değerin 0.6,0.6 değerleri verildiğinde sade grafik küçüldü
figureAxes = benimFigure2.add_axes([0.2,0.2,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
# grafiğin ortamdaki konumunu değiştirmek için ilk iki parametre
# değiştirilmelidir. Konumu değiştirir
eksen2=figure2.add_axes([0.1,0.1,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik")
eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=2,ncols=1) # tuple döndürür.
type(benimEksenler)
for eksen in benimEksenler: #eksenler bir np arraydır.
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    eksen.set_ylabel("Y ekseni")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2) # tuple döndürür.
type(benimEksenler)
for eksen in benimEksenler: #eksenler bir np arraydır.
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    eksen.set_ylabel("Y ekseni")
plt.tight_layout() #iki grafik arası mesafeyi arttırır
```
